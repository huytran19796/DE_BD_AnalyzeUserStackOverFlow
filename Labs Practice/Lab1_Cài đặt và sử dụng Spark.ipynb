# 3. Tạo SPARK
- Cần khởi tạo 1 SparkSesson để có thể bắt đầu Spark.
- Sau khi kết nối thành công với Google Drive, bạn có thể chạy thử ứng dụng Spark đầu tiên trên Colab

```python
from pyspark import SparkConf
from pyspark.sql import SparkSession

config = SparkConf().setMaster('local').setAppName('Lab1')
spark = SparkSession.builder.config(conf = config).getOrCreate()
sc = spark.sparkContext
```

```python
import collections
DATASET_PATH = '/content/gdrive/MyDrive/ml-100k/u.data'

lines = sc.textFile(DATASET_PATH)
ratings = lines.map(lambda x:x.split()[2])
result = ratings.countByValue()

sorted_result = collections.OrderedDict(sorted(result.items()))
for key, value in sorted_result.items():
  print('%s %i' % (key, value))
```
